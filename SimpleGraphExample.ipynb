```python
import netCDF4
import urllib
from mpl_toolkits.basemap import Basemap

url = ("http://n5eil01u.ecs.nsidc.org/opendap/SMAP/SPL1BTB.003/2015.05.23/"
       "SMAP_L1B_TB_01630_D_20150523T004154_R13080_001.h5.nc")

dataset = netCDF4.Dataset("SMAP_L1B_TB_01630_D_20150523T004154_R13080_001.h5.nc")

for var in dataset.variables:
    print(var)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

sea_ice = dataset["_Brightness_Temperature_sea_ice_fraction"][:,:]
lat = dataset["_Spacecraft_Data_sc_nadir_lat"][:]
lon = dataset["_Spacecraft_Data_sc_nadir_lon"][:]

plt.figure(figsize=(20,20))
plot = plt.plot(sea_ice, lat)
plt.xlabel("Sea Ice fraction")
plt.ylabel("Latitude")
plt.title("Sea Ice Fraction for latitude")
```
